```python
import sys
from pathlib import Path


def add_project_src_to_path() -> None:
    start = Path.cwd().resolve()
    candidates = [start, *start.parents]

    for base in candidates:
        for relative_path in (
            Path("services/churn_pipeline/src"),
            Path("services/churn_pipeline"),
            Path("src"),
        ):
            candidate = (base / relative_path).resolve()
            if (candidate / "churn_pipeline").exists():
                sys.path.insert(0, str(candidate))
                print(f"Added {candidate} to sys.path")
                return

    print("Could not locate churn_pipeline source directory automatically.")


add_project_src_to_path()
print(sys.executable)
print(sys.path[0])
```
